# 훈련 데이터와 데스트 데이터
* 150개 붓꽅 데이터를 두 그룹으로 나누다
* 훈련 데이터(traing set)와 데스트 데이터(hold-out-set)로 나누다
* 모델의 성능을 측정하기 위해 새 데이터 모델에 적용해야하기 떄문
* scikit-learn은 데이터셋을 섞어서 나눠주는 train_test_split 함수 제공
* 이 함수는 전체 행중 75%를 훈련세트로 뽑고, 나머지 25%는 테스트 셋
* scikit-learn에서 데이터는 대문자 X로 표시하고 레이블(붓꽃 품종)은 소문자 y로 표기
* 이는 수학 f(x)=y 에서 유래되었다. 데이터는 2차원 행렬이므로 대문자 X를 타깃(붓꽃 품종)은 1차원 배열이므로 소문자 y를 사용

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
iris_dataset['data'], iris_dataset['target'], random_state=0)

print('X_train 크기', X_train.shape)
print('y_train 크기', y_train.shape)
print('X_test 크기', X_test.shape)
print('y_test 크기', y_test.shape)
```

# KNN의 특징
* 일반적으로 KNeighbors 분류기에 중요한 매개변수는 두 개
* 데이터 포인트 사이의 거리를 재는 방법과 이웃의 수
* 거리를 재는 방법은 유클리디안 거리방식 사용
* KNN의 장점은 이해하기 쉬운 모델
* 이해하기 쉽지만 예측이 느리고 많은 특성을 처리하는 능력 부족

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

```python
X_new = np.array([[5, 2.9, 1, 0.2]])
print('X_new.shape:', X_new.shape)
```

```python
prediction = knn.predict(X_new)
print('에측:', prediction)
print('예측한 타깃의 이름:', 
     iris_dataset['target_names'][prediction])
```

# 모델평가

```python
y_pred = knn.predict(X_test)
print('테스트 세트에 대한 예측값: \n', y_pred)
```

```python
print('테스트 세트의 정확도:{:.2f}'.format(
    np.mean(y_pred == y_test)))
```

# 요약

```python
X_train, X_test, y_train, y_test = train_test_split(
iris_dataset['data'], iris_dataset['target'], random_state=0)

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)

print('테스트 세트의 정확도: {:.2f}'.format(knn.score(X_test, y_test)))
```
